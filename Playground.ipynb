```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('data/breast-aesthetics/dataset.csv')
```

```python
y = df['Overall classification'].values
for ft in df.columns.values:
    values = df[ft].values
    svalues = sorted(zip(values, y))
    sv, sy = zip(*svalues)

    left = []
    right = []

    for i in range(len(sv)):
        left.append(np.mean(sy[:i]))
        right.append(np.mean(sy[i:]))
    left = np.asarray(left)
    right = np.asarray(right)

    plt.suptitle(ft)

    plt.subplot(121)
    plt.plot((left))
    plt.subplot(122)
    plt.scatter(values, y + np.random.normal(loc=0, scale=1e-1, size=df.shape[0]))
    plt.show()
```
